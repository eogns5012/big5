```python
df = big5.loc[:, "E1" : ]
```

```python
def get_score(df, plus, minus):
    output = []
    for col in df.columns:
        if col in plus:
            output.append(df[col])
        elif col in minus:
            output.append(6 - df[col])
    return output
```

```python
plus = ['E1', 'E3', 'E5', 'E7', 'E9', 
        'A2', 'A4', 'A6', 'A8', 'A9', 'A10',
        'C1', 'C3', 'C5', 'C7', 'C9', 'C10',
        'N2', 'N4',
        'O1', 'O3', 'O5', 'O7', 'O8', 'O9', 'O10']
minus = ['E2', 'E4', 'E6', 'E8', 'E10',
         'A1', 'A3', 'A5', 'A7',
         'C2', 'C4', 'C6', 'C8',
         'N1', 'N3', 'N5', 'N6', 'N7', 'N8', 'N9', 'N10',
         'O2', 'O4', 'O6']

score = get_score(df, plus, minus)
```

```python
score
```

```python
df
```

```python
scored_df = pd.DataFrame(score)
```

```python
scored_df = scored_df.T
```

```python
scored_df
```

```python
import numpy as np
e_score = scored_df.loc[:, "E1":'E10'].sum(axis=1)
n_score = scored_df.loc[:, "N1":'N10'].sum(axis=1)
a_score = scored_df.loc[:, "A1":'A10'].sum(axis=1)
c_score = scored_df.loc[:, "C1":'C10'].sum(axis=1)
o_score = scored_df.loc[:, "O1":'O10'].sum(axis=1)
```

```python
scored_df_result = pd.concat([scored_df, e_score, n_score, a_score, c_score, o_score], axis=1)
```

```python
scored_df_result
```

```python
scored_df_result.rename(columns = {0 : 'E_SC', 1 : 'N_SC', 2 : 'A_SC', 3 : 'C_SC', 4 : 'O_SC'}, inplace = True)
```
